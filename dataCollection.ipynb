```python
#Aidan Leith
#August 31, 2024
#Used for generating requests to FPL API and storing data on all individual players, being saved to a json file

import requests
import json
import time
from datetime import datetime

#print out data
def jprint(obj):
    text = json.dumps(obj, sort_keys = True, indent = 4)
    print(text)

#get data for each player 
def fetchData(playerId):
    url = f"https://fantasy.premierleague.com/api/element-summary/{playerId}/"

    response = requests.get(url)
    if response.status_code == 200:
        return response.json()

    else:
        print("error")
        return None

#ensure data collection in file is accurate and accessible 
def checkFile():
    with open("player_data.json", "r") as file:
        data = json.load(file)

    print(json.dumps(data.get("401"), sort_keys = True, indent = 4))

#write data to data file
def dataFile(data):
    
    with open("player_data.json", "w") as file:
        json.dump(data, file, sort_keys = True, indent = 4)


#main function
def main():

    #player Ids list 
    playerIds = list(range(1, 638))
    allData = {}

    #get data for each id, add json response to dictionary of all data
    for x in playerIds:
        data = fetchData(x)
        allData[x] = data

    #only rewrite file if day is tuesday or friday
    currentDay = datetime.now().weekday()
    if currentDay in [1, 4]:
            dataFile(allData)
    
    #checkFile()

main()
```
